```python
from diffusers import AudioLDMPipeline
import torch
import scipy

repo_id = "cvssp/audioldm-s-full-v2"
pipe = AudioLDMPipeline.from_pretrained(repo_id, torch_dtype=torch.float16)
pipe = pipe.to("cuda")

prompt = "Gunshots"
negative_prompt = "Low quality, average quality"
audio = pipe(prompt, negative_prompt=negative_prompt, num_inference_steps=10, audio_length_in_s=5.0).audios[0]

# save the audio sample as a .wav file
scipy.io.wavfile.write("gun.wav", rate=16000, data=audio)
```

```python
import scipy
import torch
from diffusers import AudioLDMPipeline

repo_id = "cvssp/audioldm-m-full"
pipe = AudioLDMPipeline.from_pretrained(repo_id, torch_dtype=torch.float16)
pipe = pipe.to("cuda")

```

```python
def generateAudios(dataset="", prompt_file="", startIndex=0, num_prompts=10):
  directory = colab_path + '/' + dataset
  if not os.path.exists(directory):
      os.makedirs(directory)

  negative_prompt = "Low quality"

  with open(prompt_file) as file:
      prompts = [line.rstrip() for line in file]
  prompts = prompts[startIndex:startIndex + num_prompts]
  print(prompts)
  generator = torch.Generator("cuda").manual_seed(0)

  for i, prompt in enumerate(prompts):
    # set the seed for generator

    # run the generation
    audio = pipe(
        prompt,
        negative_prompt=negative_prompt,
        num_inference_steps=200,
        audio_length_in_s=5,
        guidance_scale=3.5,
        num_waveforms_per_prompt=3,
        generator=generator,
    ).audios

    # for j in range(0, 3 // 3):
    #   # save the best audio sample (index 0) as a .wav file
    scipy.io.wavfile.write(directory + '/'+f"{i+startIndex}_0.wav", rate=16000, data=audio[0])
```

```python
generateAudios(dataset="dog", prompt_file="dog.txt", startIndex=0, num_prompts=50)
```
